```python
#Question 1 -> python

# Start with a as 0
a = 0

# Function b() will update a, using it as a global variable
def b():
    global a
    a = c(a)  # Calls c(a) to add 2 to a

# Function c() just returns a plus 2
def c(a):
    return a + 2

# Call b() three times to increase a by 2 each time
b()  # After first call, a is 2
b()  # After second call, a is 4
b()  # After third call, a is 6

# Showed the final value of a
print(a)  # Should print 6
```

```python
#Question 3 -> python

# Define a Marsupial class
class Marsupial:
    def __init__(self):
        # Start with an empty pouch
        self.pouch = []
    
    def put_in_pouch(self, item):
        # Add an item to the pouch
        self.pouch.append(item)
    
    def pouch_contents(self):
        # Return everything in the pouch
        return self.pouch

# Define a Kangaroo class as a type of Marsupial
class Kangaroo(Marsupial):
    def __init__(self, pos_x, pos_y):
        # Set up the pouch and start location
        super().__init__()
        self.pos_x = pos_x
        self.pos_y = pos_y

    def jump(self, step_x, step_y):
        # Move the kangaroo by step_x and step_y
        self.pos_x += step_x
        self.pos_y += step_y

    def __str__(self):
        # Show where the kangaroo is
        return f"I am a Kangaroo located at coordinates ({self.pos_x},{self.pos_y})"

# Test code
k = Kangaroo(0, 0)  # Start the kangaroo at (0,0)
print(k)  # Should show: I am a Kangaroo located at coordinates (0,0)

k.put_in_pouch('doll')
k.put_in_pouch('firetruck')
k.put_in_pouch('kitten')
print(k.pouch_contents())  # Should show: ['doll', 'firetruck', 'kitten']

k.jump(1, 0)
k.jump(1, 0)
k.jump(1, 0)
print(k)  # Should show: I am a Kangaroo located at coordinates (3,0)
```

```python
# Question 4 -> python

def collatz(n):
    # Print the current value of n
    print(n)
    
    # Base case: stop when n is 1
    if n == 1:
        return
    # If the current number is even, calculate the next number by dividing by 2
    # Then call collatz() again with this new number
    elif n % 2 == 0:
        collatz(n // 2)
    # If n is odd, call collatz with 3 * n + 1
    # Call collatz() again with this new number to continue the sequence
    else:
        collatz(3 * n + 1)

# Testing the function
collatz(10)  
```

```python
# Question 5 -> python

def binary(n):
    # If n is 0 or 1, just print it since that's already the binary form
    if n < 2:
        print(n, end='')  # Print without a new line to keep building the binary number
    else:
        # First, call binary() with n divided by 2 to handle the next part
        binary(n // 2)
        # Then print the remainder of n divided by 2, which is the next binary digit
        print(n % 2, end='')

# Testing the function
print(">>> binary(0)")
binary(0)  
print("\n>>> binary(1)")
binary(1) 
print("\n>>> binary(2)")
binary(2)
print("\n>>> binary(3)")
binary(3)  
print("\n>>> binary(9)")
binary(9)  

# Referenced this StackOverflow post for converting an integer to binary in Python:
# https://stackoverflow.com/questions/10411085/converting-integer-to-binary-in-python
```

```python
# Question 6-> python

from html.parser import HTMLParser

# Define the HeadingParser class to parse HTML headings
class HeadingParser(HTMLParser):
    def __init__(self):
        # Initialize the parser and set up a way to keep track of the heading level
        super().__init__()
        self.heading_level = None  # This will store the level of the current heading

    def handle_starttag(self, tag, attrs):
        # Check if the tag we’ve hit is one of the heading tags (h1, h2, etc.)
        if tag in ('h1', 'h2', 'h3', 'h4', 'h5', 'h6'):
            # If it is, set the heading level based on the tag number (e.g., h1 = 1, h2 = 2)
            # This lets us know how much to indent the heading when we print it
            self.heading_level = int(tag[1])

    def handle_data(self, data):
        # If there’s data inside a heading tag, we’ll print it with indentation
        if self.heading_level:
            # Multiply spaces by (heading level - 1) to set the right indentation
            print(' ' * (self.heading_level - 1) + data)

    def handle_endtag(self, tag):
        # Once we reach the end of a heading tag, we clear the level
        # This way, we only print data while we’re actually inside a heading
        if tag in ('h1', 'h2', 'h3', 'h4', 'h5', 'h6'):
            self.heading_level = None

# simple HTML string here with different levels of headings.
html_content = """
<html>
<head><title>Test Document</title></head>
<body>
<h1>W3C Mission</h1>
<h2>Principles</h2>
<h3>Openness</h3>
<h3>Accessibility</h3>
<h2>Collaboration</h2>
</body>
</html>
"""

# Creating a HeadingParser instance to handle the HTML parsing
parser = HeadingParser()
parser.feed(html_content)

# Referenced this StackOverflow post for parsing HTML and extracting elements with HTMLParser:
# https://stackoverflow.com/questions/3075550/how-can-i-get-href-links-from-html-using-python
```

```python
# Question 7 -> python

import requests
from bs4 import BeautifulSoup
from urllib.parse import urljoin

def webdir(url, depth, indent):
    # Print the current URL with indentation based on depth
    print(' ' * indent + url)
    
    # Stop if we’ve reached the max depth
    if depth == 0:
        return

    # Get the HTML content of the page
    response = requests.get(url)
    soup = BeautifulSoup(response.text, 'html.parser')

    # Find all the links on the page
    for link in soup.find_all('a', href=True):
        href = link.get('href')
        # Turn any relative links into full URLs
        full_url = urljoin(url, href)
        # Call webdir() again for each link, reducing depth and adding indentation
        webdir(full_url, depth - 1, indent + 2)
        
#running the function with an URL and settings:
webdir('http://reed.cs.depaul.edu/lperkovic/csc242/test1.html', 2, 0)

# Referenced this StackOverflow post for extracting links with BeautifulSoup and implementing a recursive web crawler:
# https://stackoverflow.com/questions/1936466/beautifulsoup-grab-visible-webpage-text
```

```python
3Question 8 -> python

import pandas as pd
import sqlite3

# Load the Excel file into a DataFrame
file_path = "database.xlsx"  # Ensure this file is in the same directory or specify the full path
excel_data = pd.read_excel(file_path)

# Connect to an in-memory SQLite database
conn = sqlite3.connect(":memory:")

# Load the DataFrame to the SQLite database
excel_data.to_sql("weather_data", conn, index=False, if_exists="replace")

# SQL Queries for Question 8

# a) All the temperature data
query_a = "SELECT `Temperature (C)` FROM weather_data;"
print("All temperature data:")
print(pd.read_sql_query(query_a, conn))

# b) All the cities, without repetition
query_b = "SELECT DISTINCT City FROM weather_data;"
print("\nAll unique cities:")
print(pd.read_sql_query(query_b, conn))

# c) All the records for India
query_c = "SELECT * FROM weather_data WHERE Country = 'India';"
print("\nAll records for India:")
print(pd.read_sql_query(query_c, conn))

# d) All the Fall records
query_d = "SELECT * FROM weather_data WHERE Season = 'Fall';"
print("\nAll Fall records:")
print(pd.read_sql_query(query_d, conn))

# e) City, country, and season with rainfall between 200 and 400 mm
query_e = "SELECT City, Country, Season FROM weather_data WHERE `Rainfall (mm)` BETWEEN 200 AND 400;"
print("\nCity, country, and season with rainfall between 200 and 400 mm:")
print(pd.read_sql_query(query_e, conn))

# f) City and country with Fall temperature above 20°C, in ascending order of temperature
query_f = """
SELECT City, Country FROM weather_data 
WHERE Season = 'Fall' AND `Temperature (C)` > 20 
ORDER BY `Temperature (C)` ASC;
"""
print("\nCity and country with Fall temperature above 20°C:")
print(pd.read_sql_query(query_f, conn))

# g) Total annual rainfall for Cairo
query_g = "SELECT SUM(`Rainfall (mm)`) AS Total_Annual_Rainfall FROM weather_data WHERE City = 'Cairo';"
print("\nTotal annual rainfall for Cairo:")
print(pd.read_sql_query(query_g, conn))

# h) Total rainfall for each season
query_h = "SELECT Season, SUM(`Rainfall (mm)`) AS Total_Rainfall FROM weather_data GROUP BY Season;"
print("\nTotal rainfall for each season:")
print(pd.read_sql_query(query_h, conn))

# Close the connection
conn.close()
```
